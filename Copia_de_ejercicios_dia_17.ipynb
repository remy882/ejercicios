5X + 2Y = 1

-3x + 3y = 5

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.array([[5, 2], [-3, 3]])
b = np.array([1, 5])
x = np.linalg.solve(a, b)

puntos_x = [0, x[0]]
puntos_y = [0, x[1]]

plt.figure(figsize=(8, 6))

plt.plot(puntos_x, puntos_y, 'o--', color='blue', label='Puntos de la solución')

# Ajustamos los límites de los ejes para que el gráfico sea más legible
plt.xlim(-1, 0.5)
plt.ylim(0, 2.5)

# Añadimos las anotaciones
plt.annotate(f'Inicio\n(0,0)', xy=(0, 0), xytext=(-0.5, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate(f'Solución\n({x[0]:.2f}, {x[1]:.2f})', xy=(x[0], x[1]),
             xytext=(-0.5, 2),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Representación de la solución del sistema')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parte 1: Resolver el sistema de ecuaciones lineales
a = np.array([[5, -1], [-2, 4]])
b = np.array([3, 12])
x = np.linalg.solve(a, b)

# El resultado de la solución es:
# x[0] es el valor para la primera variable
# x[1] es el valor para la segunda variable
print(f"La solución al sistema de ecuaciones es x = {x[0]:.2f} y y = {x[1]:.2f}")

# Parte 2: Crear un gráfico
# Supongamos que queremos graficar una línea simple como y = m*x + c
# Usaremos los valores de la solución para definir nuestra ecuación
# Por ejemplo, podemos usar y = x[0]*x + x[1]
x_valores = np.linspace(-10, 10, 100)
y_valores = x[0] * x_valores + x[1]

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_valores, y_valores, color='red', label=f'y = {x[0]:.2f}x + {x[1]:.2f}')

# Añadir elementos al gráfico
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de la solución del sistema de ecuaciones')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parte 1: Resolver el sistema de ecuaciones lineales
a = np.array([[5, -1], [-2, 4]])
b = np.array([3, 12])
x = np.linalg.solve(a, b)

# Los puntos que queremos graficar son (0,0) y la solución (x[0], x[1])
puntos_x = [0, x[0]]  # Los valores de x para los dos puntos
puntos_y = [0, x[1]]  # Los valores de y para los dos puntos

# Parte 2: Crear un gráfico que solo muestre esos puntos
plt.figure(figsize=(8, 6))

# Trazamos los puntos
# El argumento 'o--' crea una línea de puntos con círculos en los extremos
plt.plot(puntos_x, puntos_y, 'o--', color='blue', label='Puntos de la solución')

# Añadir etiquetas para cada punto para mayor claridad
plt.annotate(f'Inicio\n(0,0)', xy=(0, 0), xytext=(0, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate(f'Solución\n({x[0]:.2f}, {x[1]:.2f})', xy=(x[0], x[1]), xytext=(x[0] + 0.5, x[1] + 1),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Añadir elementos al gráfico
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Representación de la solución del sistema')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5) # Eje x
plt.axvline(0, color='black', linewidth=0.5) # Eje y
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los valores para el eje X
x_vals = np.linspace(-10, 10, 400)

# Ecuación 1: -x + 2y = 4  -->  y = 0.5x + 2
y1_vals = 0.5 * x_vals + 2

# Ecuación 2: 2x - 4y = 3  -->  y = 0.5x - 0.75
y2_vals = 0.5 * x_vals - 0.75

# Crear el gráfico
plt.figure(figsize=(8, 6))

# Trazar la primera línea
plt.plot(x_vals, y1_vals, label='-x + 2y = 4', color='blue')

# Trazar la segunda línea
plt.plot(x_vals, y2_vals, label='2x - 4y = 3', color='red')

# Añadir los elementos del gráfico
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Representación de un Sistema de Ecuaciones Inconsistente')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la matriz de coeficientes y el vector de términos independientes
a = np.array([[2, 3], [3, 2]])
b = np.array([1, 4])
x = np.linalg.solve(a, b)

#graficar los puntos
puntos_x = [0, x[0]]
puntos_y = [0, x[1]]

#crear los garficos
plt.figure(figsize=(8, 6))

#Trazar la linea
plt.plot(puntos_x, puntos_y, 'o--', color='blue', label='Solución del sistema')
# Añadir la solución como texto en el gráfico
solucion_texto = f'La solución es:\n x = {x[0]:.2f}\n y = {x[1]:.2f}'
plt.text(0.7, 0.9, solucion_texto, transform=plt.gca().transAxes,
         bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.5))
# Añadir los elementos en el grafico
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Representación de la solución')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
#mostramos
plt.show()
```

4x -3y = 5

-8x +6y = 10

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los valores para el eje X
x_vals = np.linspace(-10, 10, 400)

# Ecuación 1: 4x - 3y = 5  -->  y = (4/3)x - 5/3
y1_vals = (4/3) * x_vals - 5/3

# Ecuación 2: -8x + 6y = 10  -->  y = (8/6)x + 10/6  -->  y = (4/3)x + 5/3
y2_vals = (4/3) * x_vals + 5/3

# Crear el gráfico
plt.figure(figsize=(8, 6))

# Trazar la primera línea
plt.plot(x_vals, y1_vals, label='4x - 3y = 5', color='blue')

# Trazar la segunda línea
plt.plot(x_vals, y2_vals, label='-8x + 6y = 10', color='red')

# Añadir los elementos del gráfico
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Representación de un Sistema de Ecuaciones Sin Solución')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Mostrar el gráfico
plt.show()
```
